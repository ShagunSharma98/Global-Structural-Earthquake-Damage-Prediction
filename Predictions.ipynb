```python
cols_1=n.concatenate([n.arange(178),n.arange(179,285),n.arange(286,332),n.arange(333,440)],axis=0)

missing_joints=['Joint179','Joint286','Joint333']
```

```python
from keras import models 
Network_1 = []

for key, index in zip(cols_1, range(437)):
  Network_1.append(models.load_model('1D_CNN_Models/Model_Joint_'+str(key)+'.h5'))
```

## Preparing Data For Predictions

```python
path_1="drive/MyDrive/Structural Damage Prediction/SRINAGAR/"

g=[0.1,0.3,0.5,1,2]
g0=[3]
```

```python
ud_srinagar_x={}

file_list_1=glob.glob(path_1+"srinagar ACC X *(U).csv")

for file_path in file_list_1:
    ud_srinagar_x[file_path[file_path.rfind(' ')+1:-7]]=p.DataFrame(p.read_csv(file_path))
    for column in ud_srinagar_x[file_path[file_path.rfind(' ')+1:-7]].columns:
        if column in missing_joints:
            ud_srinagar_x[file_path[file_path.rfind(' ')+1:-7]]=ud_srinagar_x[file_path[file_path.rfind(' ')+1:-7]].drop(column,axis=1) 

ud_srinagar_y={}

file_list_2=glob.glob(path_1+"srinagar ACC Y *(U).csv")

for file_path in file_list_2:
    ud_srinagar_y[file_path[file_path.rfind(' ')+1:-7]]=p.DataFrame(p.read_csv(file_path))
    for column in ud_srinagar_y[file_path[file_path.rfind(' ')+1:-7]].columns:
        if column in missing_joints:
            ud_srinagar_y[file_path[file_path.rfind(' ')+1:-7]]=ud_srinagar_y[file_path[file_path.rfind(' ')+1:-7]].drop(column,axis=1) 
```

```python
ud_srinagar_x['0.1']=ud_srinagar_x['0.1'].drop('Unnamed: 0',axis=1)
ud_srinagar_x['0.3']=ud_srinagar_x['0.3'].drop('Unnamed: 0',axis=1)
ud_srinagar_x['0.5']=ud_srinagar_x['0.5'].drop('Unnamed: 0',axis=1)
# ud_srinagar_x['0.7']=ud_srinagar_x['0.7'].drop('Unnamed: 0',axis=1)
ud_srinagar_x['1']=ud_srinagar_y['1'].drop('Unnamed: 0',axis=1)
ud_srinagar_x['2']=ud_srinagar_x['2'].drop('Unnamed: 0',axis=1)

ud_srinagar_y['0.1']=ud_srinagar_y['0.1'].drop('Unnamed: 0',axis=1)
ud_srinagar_y['0.3']=ud_srinagar_y['0.3'].drop('Unnamed: 0',axis=1)
ud_srinagar_y['0.5']=ud_srinagar_y['0.5'].drop('Unnamed: 0',axis=1)
# ud_srinagar_y['0.7']=ud_srinagar_y['0.7'].drop('Unnamed: 0',axis=1)
ud_srinagar_y['1']=ud_srinagar_y['1'].drop('Unnamed: 0',axis=1)
ud_srinagar_y['2']=ud_srinagar_y['2'].drop('Unnamed: 0',axis=1)

ud_srinagar_x['0.1'].columns=cols_1
ud_srinagar_x['0.3'].columns=cols_1
ud_srinagar_x['0.5'].columns=cols_1
# ud_srinagar_x['0.7'].columns=cols_1
ud_srinagar_x['1'].columns=cols_1
ud_srinagar_x['2'].columns=cols_1

ud_srinagar_y['0.1'].columns=cols_1
ud_srinagar_y['0.3'].columns=cols_1
ud_srinagar_y['0.5'].columns=cols_1
# ud_srinagar_y['0.7'].columns=cols_1
ud_srinagar_y['1'].columns=cols_1
ud_srinagar_y['2'].columns=cols_1
```

```python
ud_srinagar={}

i=0
while(i<5):
    ud_srinagar[str(g[i])]=p.DataFrame(p.concat((ud_srinagar_x[str(g[i])],ud_srinagar_y[str(g[i])]),axis=0))
    ud_srinagar[str(g[i])].reset_index(drop=True,inplace=True)
    i+=1
```

```python
d_srinagar_x={}

file_list_3=glob.glob(path_1+"srinagar ACC X *(D).csv")

for file_path in file_list_3:
    d_srinagar_x[file_path[file_path.rfind(' ')+1:-7]]=p.DataFrame(p.read_csv(file_path))
    for column in d_srinagar_x[file_path[file_path.rfind(' ')+1:-7]].columns:
        if column in missing_joints:
            d_srinagar_x[file_path[file_path.rfind(' ')+1:-7]]=d_srinagar_x[file_path[file_path.rfind(' ')+1:-7]].drop(column,axis=1) 

d_srinagar_y={}

file_list_4=glob.glob(path_1+"srinagar ACC Y *(D).csv")

for file_path in file_list_4:
    d_srinagar_y[file_path[file_path.rfind(' ')+1:-7]]=p.DataFrame(p.read_csv(file_path))
    for column in d_srinagar_y[file_path[file_path.rfind(' ')+1:-7]].columns:
        if column in missing_joints:
            d_srinagar_y[file_path[file_path.rfind(' ')+1:-7]]=d_srinagar_y[file_path[file_path.rfind(' ')+1:-7]].drop(column,axis=1) 
```

```python
d_srinagar_x['3']=d_srinagar_x['3'].drop('Unnamed: 0',axis=1)

d_srinagar_y['3']=d_srinagar_y['3'].drop('Unnamed: 0',axis=1)

d_srinagar_x['3'].columns=cols_1

d_srinagar_y['3'].columns=cols_1
```

```python
d_srinagar={}

i=0
while(i<1):
    d_srinagar[str(g0[i])]=p.DataFrame(p.concat((d_srinagar_x[str(g0[i])],d_srinagar_y[str(g0[i])]),axis=0))
    d_srinagar[str(g0[i])].reset_index(drop=True,inplace=True)
    i+=1
```

```python
UD_test={}

for i in cols_1:
    UD_test[i]=p.DataFrame()
    for g in g:
        UD_test[i]=UD_test[i].append(ud_srinagar[str(g)][i])
    UD_test[i].reset_index(drop=True,inplace=True)
    
D_test={}

for i in cols_1:
    D_test[i]=p.DataFrame()
    for g in g0:
        D_test[i]=D_test[i].append(d_srinagar[str(g)][i])
    D_test[i].reset_index(drop=True,inplace=True)
```

```python
d_actual={}

for key in cols_1:
    d_actual[key]=p.Series()
    for i in range(D_test[key].shape[0]):
        d_actual[key].loc[i]=n.float64(0)
        
ud_actual={}

for key in cols_1:
    ud_actual[key]=p.Series()
    for i in range(UD_test[key].shape[0]):
        ud_actual[key].loc[i]=n.float64(1)
```

```python
test_data={}

for key in cols_1:
    test_data[key]=p.DataFrame()
    test_data[key]=p.concat((D_test[key],UD_test[key]),axis=0)
    test_data[key].reset_index(drop=True,inplace=True)

test_act_out={}

for key in cols_1:
    test_act_out[key]=p.Series()
    test_act_out[key]=p.concat((d_actual[key],ud_actual[key]),axis=0)
    test_act_out[key].reset_index(drop=True,inplace=True)
```

```python
test_Input = {}
test_Output = {}

for key in cols_1:
    test_Input[key]=n.reshape(n.asarray(test_data[key]),(test_data[key].shape[0],4001,2))
    test_Output[key]=n.reshape(n.asarray(test_act_out[key]),(test_act_out[key].shape[0],1))
```

```python
preds={}

for key,index in zip(cols_1, range(437)):
    # preds[key] = Network_1[index].predict(test_Input[key])
    preds[key] = Network_1[index](test_Input[key])    # To avoid tracing warnings
```

## Probabilities of Damage

```python
scores_1=0
efficiency_1=0
for key,index in zip(cols_1,range(437)):
    scores_1=scores_1+Network_1[index].evaluate(test_Input[key],test_Output[key])[1]
efficiency_1=(scores_1/n.shape(Network_1)[0])*100
```

```python
print("\nEfficiency Obtained from 1D CNN Network on Test Dataset is: "+ str(efficiency_1)+"%")
```
